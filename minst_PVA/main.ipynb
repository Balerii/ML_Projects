```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
import sklearn as sk
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train=np.concatenate((x_test,x_train),axis=0)
y_train=np.concatenate((y_test,y_train),axis=0)
#In this task I don't need test dataset so I can combine test end train data
```

```python
x_train=x_train.reshape(-1,28*28)
```

```python
pca=sk.decomposition.PCA(n_components=28*28)
pca.fit(x_train)
```

```python
rat=pca.explained_variance_ratio_
plt.plot(np.arange(len(rat)),np.cumsum(rat))
```

how much "information" contains first n coordinates

```python
pca=sk.decomposition.PCA(n_components=2)
pca.fit(x_train)
x=pca.transform(x_train)
```

```python
plt.figure(figsize=(14, 14))
for i in range(10):
    plt.scatter(*x[y_train==i].T, label=i)
plt.legend()
```

Scatter plot of projection of dataset on first two most "informative" coordinates.

We can see that same numbers are grouped

```python
y_train[0],y_train[1]
```

We can create "new numbers" if take mean on the first few coordinates. For example let's combine seven and two

```python
tweven=x[0]*0.5+x[1]*0.5
inv=pca.inverse_transform(tweven).reshape(28,28)
plt.imshow(inv)
```

Too blurry. Let's take the first 100 coordinates
