```
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
```

```
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0 #масштабируем 
```

```
def create_model(n_relu=0, n_sigmoid = 0, n_softmax=0):
    model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    ])

    for i in range(n_relu) :
        model.add(tf.keras.layers.Dense(64, activation='relu'))
    
    for i in range(n_sigmoid) :
        model.add(tf.keras.layers.Dense(64, activation='sigmoid'))
    
    for i in range(n_softmax) :
        model.add(tf.keras.layers.Dense(64, activation='softmax'))
    
    return model

def train_model(model, epoch = 1):
  model.compile(optimizer='RMSprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

  model.fit(x_train, y_train,batch_size=32, epochs=epoch, shuffle=False)
  return model.evaluate(x_test, y_test, batch_size=32)


def create_data_table(*argv) :
  data = []
  for ar in argv:
    for elem in ar:
      tmp = []
      for m in elem:
        tmp.append(round(m[1], 4))
      data.append(tmp)
  return data
```

```
#создаем модели
m_relu = []
m_sigmoid = []
m_softmax = []
for i in range(1, 11):
  
  m_relu.append(create_model(i, 0, 0))
  m_sigmoid.append(create_model(0, i, 0))
  m_softmax.append(create_model(0, 0, i))

#обучаем их с эпохами 1 и 5
tr_relu = []
tr_sigmoid = []
tr_softmax = []
for epoch in range(1, 6, 2):
  print("epoch = ", epoch)
  tmp_r = []
  tmp_s = []
  tmp_smax = []
  for i in range(10):
    tmp_r.append(train_model(m_relu[i], epoch))
    tmp_s.append(train_model(m_sigmoid[i], epoch))
    tmp_smax.append(train_model(m_softmax[i], epoch))
  tr_relu.append(tmp_r)
  tr_sigmoid.append(tmp_s)
  tr_softmax.append(tmp_smax)

```

WARNING:tensorflow:Callbacks method `on_test_batch_end` is slow compared to the batch time

Іноді може спливати помилка при тренуванні з активації relu (погано підходить для цього завдання), швидше за все це пов'язано з малою кількістю батчей або **епох**.
(вспливає приблизно через раз і лише на першій епосі)

1. Найбільш ефективним з активацій є **sigmoid**, 
найгірший **relu** - отже він погано підходить для навчання з данних MNIST
2. Для кожної з активацій є свій ліміт слоїв після якого вони майже перестають змінювати точність (~15%), але цей ліміт не завжди однаковий при різних навчаннях. Я припускаю що активація яка погано підходить, зі збільшенням слоїв тільки зменшуе свою точність( це relu, softmax).
3. Зі збільшенням епох - збільшуєтся точність навчання моделі.

**Найкраща точність досягаеться у sigmoid з 3 слоями та 5 епохами**
